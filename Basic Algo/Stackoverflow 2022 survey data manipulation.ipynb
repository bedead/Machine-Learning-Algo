```python
import pandas as pd
df = pd.read_csv('database/survey_results_public.csv', index_col=0)
schema_df = pd.read_csv("database/survey_results_schema.csv", index_col=0)
```

```python
# dropping unnecessary columns

drop_col = ['CodingActivities',
           'LearnCodeCoursesCert',
           'TrueFalse_1',
           'TrueFalse_2',
           'TrueFalse_3',
           'SurveyLength',
           'SurveyEase',]

df.drop(drop_col ,axis=1)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib
# Import seaborn
import seaborn as sns
%matplotlib inline

# setting defaults
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

# same data which we extracted above 
sal = df['ConvertedCompYearly']
# removing all the NaN values
sal.dropna(inplace=True)

plt.plot(sal)
plt.xlabel('Salary')
plt.show()
```

```python
# ploting top 10 countries
# no. of unique country in data
df.Country.nunique()
```

```python
top_countries = df.Country.value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.title("Where do you live?")
sns.barplot(x=top_countries.index, y=top_countries);
```

```python
# extracting complete colunm
yearcode = df['YearsCodePro']
```

```python
yearcode.dropna(inplace=True)

plt.plot(yearcode, 'y^')
plt.ylabel('Years of Work')
plt.show()
```

```python
# ploting scatter graph
plt.scatter(sal, np.arange(0,len(sal)))
plt.ylabel("Salary")
plt.show()
```

```python
import matplotlib.patches as mpatches

plt.scatter(yearcode, np.arange(0, yearcode.shape[0] ), c='g')
plt.xlabel("Years")
green_patch = mpatches.Patch(color='green', label='Green data')
plt.legend(handles=[green_patch])
plt.show()
```
